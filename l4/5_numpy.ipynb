## Getting The Data

```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

we will use a logo image

```python
img_url = "https://pbs.twimg.com/profile_images/1517584461017370624/DF3DpXUW_400x400.jpg"
logo = imageio.imread(img_url)
print(type(logo))
print(logo.shape)
```

```python
plt.imshow(logo)
```

```python
# plot three channels in a row using matplotlib
fig, axes = plt.subplots(1, 3, figsize=(10, 5))
for i, ax in enumerate(axes):
    ax.imshow(logo[:, :, i], cmap=plt.cm.gray)
    ax.set_title(f"Channel {i}")
```

```python
# plot the logo in grayscale
plt.imshow(logo.mean(axis=2), cmap=plt.cm.gray)
```

```python
# plot the logo with inverted colors
plt.imshow(255 - logo.mean(axis=2), cmap=plt.cm.gray)
```

```python
# mirror the logo horizontally
plt.imshow(logo[:, ::-1])
```

```python
# crop the logo 
plt.imshow(logo[50:150, 50:150])
```

```python
# add some noise to the logo
plt.imshow(logo + np.random.randn(*logo.shape) * 255)
```

```python
# rotate the logo 30 degrees
plt.imshow(np.rot90(logo, 1))
```

```python
# add 3 logos together
plt.imshow(np.concatenate([logo, logo, logo], axis=1))
```
