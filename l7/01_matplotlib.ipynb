```python
# an example line chart
import  numpy  as  np
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(x), np.sin(x)
plt.plot(x, C, color="blue", linewidth=2.5, linestyle="dashdot", label="cosine")
plt.plot(x, S, color="red", linewidth=2.5, linestyle="-", label="sine")
plt.title("Cosine and Sine")
plt.legend(loc='upper left')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
              [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],
                [r'-1', r'$0$', r'$+1$'])
plt.show()
```

```python
# an example bar chart 
import  numpy  as  np
n = 12
X = np.arange(n)
Y1 = (1 - X/float(n)) * np.random.exponential(0.5, n)
Y2 = (1 - X/float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')
for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va= 'bottom')
plt.ylim(-1.25,+1.25)
plt.show()
```

```python
# an example scatter plot
import  numpy  as  np
n = 1024
X = np.random.normal(0,2,n)
Y = np.random.normal(0,2,n)
T = np.arctan2(Y,X)
plt.scatter(X, Y, c=T, s=75, alpha=.5)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
# an example distribution plot
import  numpy  as  np
import  matplotlib.pyplot  as  plt
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
# an example contour plot
import  numpy  as  np
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)
plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap=plt.cm.hot)
C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
plt.clabel(C, inline=1, fontsize=10)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
# an example subplot with shared axis
import  numpy  as  np
import  matplotlib.pyplot  as  plt
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo')
plt.plot(t2, f(t2), 'k')
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
# an example horizontal stacked bar chart
import  numpy  as  np
import  matplotlib.pyplot  as  plt
n = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(n)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence
p1 = plt.bar(ind, menMeans, width, yerr=menStd, color='b', label='men')
p2 = plt.bar(ind, womenMeans, width, bottom=menMeans, yerr=womenStd, color='r', label='women')
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.legend()
plt.title('Scores by gender over groups')   
plt.savefig('foo.png')
plt.savefig('foo.jpg')
plt.savefig('foo.pdf')
plt.savefig('foo.svg')
plt.show()
```

```python
# an example Scores by gender over groups figure with multiple axes with shared y-axis
import  numpy  as  np
import  matplotlib.pyplot  as  plt
n = 5
menMeans = (30, 45, 40, 45, 37)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2) 
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(n)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence
fig, axes = plt.subplots(nrows=1, ncols=2, sharey=True)
axes[0].bar(ind, menMeans, width, yerr=menStd, color='b', label='men')
axes[1].bar(ind, womenMeans, width, yerr=womenStd, color='r', label='women')
axes[0].set_xlabel('Groups')
axes[0].set_ylabel('Scores')
axes[1].set_xlabel('Groups')
axes[1].set_ylabel('Scores')
axes[0].legend()
axes[1].legend()
fig.show()

```
