A [data set](http://konect.cc/networks/moreno_lesmis/) with undirected network contains co-occurances of characters in Victor Hugo's novel 'Les Misérables'. 

```python
import networkx as nx
G = nx.les_miserables_graph()
```

```python
G.number_of_nodes()
```

```python
G.number_of_edges()
```

```python
list(G.nodes(data=True))[:5]
```

```python
list(G.edges(data=True))[:5]
```

```python
nx.draw(G, with_labels=True)
```

```python
layout = nx.layout.spiral_layout(G)
list(layout.items())[:5]
```

```python
nx.draw(
    G, nx.layout.fruchterman_reingold_layout(G), with_labels=True, 
    node_size=100, node_color='lightblue', 
    font_size=8, font_color='black', 
    edge_color='gray', width=1,

)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(20, 20))

nx.draw(
    G, nx.layout.fruchterman_reingold_layout(G), with_labels=True, 
    node_size=1000, node_color='lightblue', 
    font_size=20, font_color='black', 
    edge_color='gray', width=5,

)
```

```python
layout = nx.layout.fruchterman_reingold_layout(G)
list(layout.items())[:5]
```

```python
from bokeh.palettes import Category20_20
from bokeh.plotting import figure, from_networkx, show

p = figure(x_range=(-2, 2), y_range=(-2, 2),
           x_axis_location=None, y_axis_location=None,
           tools="hover", tooltips="index: @index")
p.grid.grid_line_color = None

graph = from_networkx(G, nx.spring_layout, scale=1.8, center=(0,0))
p.renderers.append(graph)

# Add some new columns to the node renderer data source
graph.node_renderer.data_source.data['index'] = list(range(len(G)))
graph.node_renderer.data_source.data['colors'] = Category20_20

graph.node_renderer.glyph.update(size=20, fill_color="colors")

show(p)
```

```python
nodename_mapper = {node_name: i for i, node_name in enumerate(G.nodes())}
G_relabeled = nx.relabel_nodes(G, nodename_mapper)
    
```

```python
from bokeh.palettes import Category20_20
from bokeh.plotting import figure, from_networkx, show

p = figure(x_range=(-2, 2), y_range=(-2, 2),
           x_axis_location=None, y_axis_location=None,
           tools="hover", tooltips="index: @index")
p.grid.grid_line_color = None

p.circle(x=0, y=0, size=100, color='red')
graph = from_networkx(G_relabeled, nx.spring_layout, scale=1.8, center=(0,0))
p.renderers.append(graph)

# Add some new columns to the node renderer data source
graph.node_renderer.data_source.data['index'] = list(range(len(G_relabeled)))
graph.node_renderer.data_source.data['colors'] = Category20_20

graph.node_renderer.glyph.update(size=20, fill_color="colors")

show(p)
```

```python
nodename_mapper_r = {value:key for key, value in nodename_mapper.items()}
```

```python
nx.set_node_attributes(G_relabeled, nodename_mapper_r, 'name')
```

```python
list(G_relabeled.nodes(data=True))[0]
```

```python
from bokeh.plotting import output_notebook
output_notebook()

from bokeh.plotting import figure, from_networkx, show

p = figure(x_range=(-2, 2), y_range=(-2, 2),
           x_axis_location=None, y_axis_location=None,
           tools="hover", tooltips="index: @index, name: @name")
p.grid.grid_line_color = None

graph = from_networkx(G_relabeled, nx.spring_layout, scale=1.8, center=(0,0))
p.renderers.append(graph)

# Add some new columns to the node renderer data source
graph.node_renderer.data_source.data['index'] = list(range(len(G_relabeled)))

show(p)
```
